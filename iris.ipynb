```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# df= sns.load_dataset('IRIS.csv')
# df.head()
data= pd.read_csv("IRIS.csv")
data.head()
```

```python
data['species'],categories = pd.factorize(data['species'])
data.head() 
```

```python
sns.scatterplot(data= data, x="sepal_length", y="sepal_width", hue="species")
```

```python
k_rng = range(1,10)
sse=[]
for k in k_rng:
    km= KMeans(n_clusters=k)
    km.fit(data[['petal_length','petal_width']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k_rng')
plt.ylabel("sum of squared errors")
plt.plot(k_rng, sse)
```

```python
km = KMeans(n_clusters=3, random_state=0,)
y_predicted = km.fit_predict(data[['petal_length','petal_width']])
y_predicted
```

```python
data['cluster']=y_predicted
data.head(120)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(data.species, data.cluster)
cm
```

```python
true_label= data.species
predicted_labels= data.cluster
cm= confusion_matrix(true_label,predicted_labels)
class_labels =['Setosa', 'versicolor', 'virginica']

# plot confusion matrix
plt.imshow(cm, interpolation='nearest', cmap= plt.cm.Blues)
plt.title('confusion matrix')
plt.colorbar()
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels)
plt.yticks(tick_marks, class_labels)

#  fill matrix with values
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j,i, str(cm[i])[j], ha='center', va='center', color='white')
        
plt.xlabel('predicted_label')
plt.ylabel('true label')
plt.show()
```
